# Define functions

```python
def boxwhiskerplot(dataframe, name):
    s =20
    columns_to_plot = dataframe.columns[:7]  # Exclude the last column for the special case
    data_to_plot = pd.melt(dataframe[columns_to_plot], var_name='Datasets', value_name='Values')
    plt.figure(figsize=(8,7)) 
    custom = {'DS-1': 'indianred', 'DS-2': 'khaki', 'DS-3': 'teal', 'DS-4': 'mediumpurple', 'DS-5': 'sandybrown', 'DS-6': 'darkseagreen','DS-7': 'lightsteelblue' }
    sns.boxplot(x= 'Datasets',y='Values', data=data_to_plot,
                 saturation=1, fliersize = 10,flierprops=dict(marker = "o", markersize = 8, markeredgecolor = "black", markerfacecolor="white"),
                 showmeans = True, showcaps = True, whiskerprops =dict(linestyle = 'dashed'), meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"red",'markersize':'6'}, 
                 notch = True, linewidth = 1,
                 palette = custom
                 )
    plt.xticks(rotation=0, fontsize = s)
    plt.yticks(rotation=0, fontsize = s)
    plt.ylabel(f'{name}', fontsize = s)
    plt.xlabel('Datasets', fontsize = s)
    plt.tick_params('both', length=10, width=1, which='major', top= True, right = True, direction = 'in')
    plt.tick_params('both', length=5, width=1, which='minor', top= True, right = True, direction = 'in')
    plt.savefig(f'./notebook_figures/Whisker/{name}.tiff',format='tiff', dpi=1200)
    plt.show()
```

# Ploting for Schematic

```python
def boxwhiskerplot(dataframe, name):
    s =20
    columns_to_plot = dataframe.columns[:7]  # Exclude the last column for the special case
    data_to_plot = pd.melt(dataframe[columns_to_plot], var_name='Datasets', value_name='Values')
    plt.figure(figsize=(2,2)) 
    custom = {'DS-1': 'indianred', 'DS-2': 'khaki', 'DS-3': 'teal', 'DS-4': 'mediumpurple', 'DS-5': 'sandybrown', 'DS-6': 'darkseagreen','DS-7': 'lightsteelblue' }
    sns.boxplot(x= 'Datasets',y='Values', data=data_to_plot,
                 saturation=1, fliersize = 2,flierprops=dict(marker = "o", markersize = 5, markeredgecolor = "white", markerfacecolor="black"),
                 notch = False, linewidth = 1,
                 palette = custom
                 )
    plt.xticks(rotation=0, fontsize = 0)  #['DS-1', 'DS-2', 'DS-3', 'DS-4', 'DS-5', 'DS-6', 'DS-7' ],
    plt.yticks(rotation=0, fontsize = 0)
    plt.ylabel('', fontsize = 0)
    plt.xlabel('', fontsize = 0)
    plt.tick_params( length=0, width=0, which='both', top= False, right = False, bottom = False, left = False, direction = 'in')
    plt.savefig(f'./notebook_figures/Schematic/whisker.tiff',format='tiff', dpi=1200)
    plt.show()

boxwhiskerplot(df_completedatadia, 'Diameter (mm)')
```
